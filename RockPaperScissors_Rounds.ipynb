```python
#Rock, Paper, Scissors: (multiple rounds)
import random

#initialize wins/draw
user_win = 0
comp_win = 0 
draws = 0
match = 0 # count game rounds


#make loop to count wins/draws and number of matches/rounds
while True: 
    user = input("Choose one: ('R' for rock, 'P' for paper, or 'S' for scissors) ").upper()
    choices = ["R", "P", "S"]
    computer = random.choice(choices)
    
    if user == computer:
        draws += 1
        match += 1
        print("It\'s a draw!")
    
    elif user == "R": 
        if computer == "S":
            user_win += 1
            match += 1
            print("Rock smashes scissors! Congrats, you win!")
        else: 
            comp_win += 1
            match += 1
            print("Sorry, paper covers rock. Computer wins!")
    elif user == "S":
        if computer == "P":
            user_win += 1
            match += 1
            print("Scissors cut paper! Congrats, you win!")
        else: 
            comp_win += 1
            match += 1
            print("Too bad! Rock smashes scissors! Computer wins!")
    elif user == "P":
        if computer == "R":
            user_win += 1
            match += 1
            print("Paper covers rock! You win!")
        else: 
            comp_win += 1
            match += 1
            print("Sorry, scissors cut paper! Computer wins!")

    #continue game=Y/ end game=N       
    again = input("Play again? Y/N: ").upper()
    if again != "Y":
        break

print("GAME OVER")
print(f"Total Rounds: {match}")       
print(f"You won: {user_win} time(s)")
print(f"Computer won: {comp_win} time(s)")
print(f"Ties: {draws}")
```
